```python
import keras
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.chdir("..")
```

```python
random_noise_size = 1000
howmuch_image = 20
```

```python
os.chdir("./models")
gan = keras.models.load_model("model_1900.h5")
os.chdir("..")
```

```python
gen,disc = gan.layers
```

```python
noise = tf.random.normal(shape=[howmuch_image,random_noise_size])
generated_imgs=gen(noise)

fig = plt.figure(figsize=(10,10))
for i in range(howmuch_image):
    plt.subplot(5,5,i+1)
    plt.imshow(generated_imgs[i, :, :, 0] * 127.5 + 127.5, cmap='binary')
    plt.axis('off')
os.chdir("./test_results")
length = os.listdir()
plt.savefig("test_1")
os.chdir("..")
```
